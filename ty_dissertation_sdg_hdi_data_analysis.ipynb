# Descriptive analysis

**HDI Trend line**

```python
hdi_df = df[df["series_name"] == "Human Development Index"]

# Select the columns for years 1990 to 2020
years = [str(year) for year in range(1990, 2021)]
hdi_values = hdi_df[years].values.flatten()

plt.figure(figsize=(20, 6))
# Iterate over each country and plot its trend line
for _, row in hdi_df.iterrows():
    country_name = row['country_name']
    hdi_values = row[years].values.flatten()
    plt.plot(years, hdi_values, label=country_name)

# Add labels and legend to the plot
plt.xlabel('Year')
plt.ylabel('HDI')
plt.title('Trend of Human Development Index')
plt.legend()
# Show the plot
plt.show()
```

**GDP Trend line**

```python
gdp_df = df[df["series_name"] == "GDP (constant 2015 US$)"]

# Select the columns for years 1990 to 2020
years = [str(year) for year in range(1990, 2021)]
gdp_values = gdp_df[years].values.flatten()

plt.figure(figsize=(20, 6))
# Iterate over each country and plot its trend line
for _, row in gdp_df.iterrows():
    country_name = row['country_name']
    gdp_values = row[years].values.flatten()/1000000000
    plt.plot(years, gdp_values, label=country_name)

# Add labels and legend to the plot
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('Trend of Gross Domestic Product(constant 2015 US$ Billion)')
plt.legend()
# Show the plot
plt.show()
```

# Model Building

**Feature importance for Nigeria**

**Get A Country Data**

```python
country_data = df[df['country_name'] == 'Nigeria']
# drop 'country_name','country_code'
country_data.drop(['country_name','country_code'], axis=1, inplace=True)
```

**Transpose data**

```python
country_data_transpose = country_data.transpose()
```

**set axis**

```python
# Replace column names with values from the first row
country_data_transpose = country_data_transpose.set_axis(country_data_transpose.iloc[0], axis=1)

# Drop the first row (previous column names row)
country_data_transpose = country_data_transpose[1:]
country_data_transpose
```

**Drop Human Development Index**

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

**Fill null fields with zero**

```python
country_data_transpose = country_data_transpose.fillna(0)
```

```python
# Identify and assign the features and target to a variable
X = country_data_transpose.drop(['Human Development Index'], axis=1)
y = country_data_transpose['Human Development Index']
```

**Split data to test and train data**

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
y_test.dtype
```

```python
# Now let's do random forest regression
rf_reg = RandomForestRegressor().fit(X_train, y_train)
rf_predicted = rf_reg.predict(X_test)
print(f"Random Forest Score: {rf_reg.score(X_train, y_train)}")
print(f"Mean Absolute Error for RF: {mean_absolute_error(y_test, rf_predicted)}")
print('Test MSE score:', mean_squared_error(y_test, rf_predicted))
print('Test R2 score:', r2_score(y_test, rf_predicted))
print('Test RMSE score:', np.sqrt(mean_squared_error(y_test, rf_predicted)))
```

```python
def make_plot(truth, prediction):
  plt.plot(truth, color="red", label="truth")
  plt.plot(prediction, color="blue", label="predicted")
  plt.legend()
  plt.grid()
  plt.title("Comparing truth and predicted regression values")
  plt.tight_layout()
  plt.show()
```

```python
make_plot(y_test, rf_predicted)
```

**Feature importance**

```python
rf_importances = rf_reg.feature_importances_
rf_reg.feature_names_in_
ranks_and_features = zip(rf_importances, rf_reg.feature_names_in_)
ranks_and_features = sorted(ranks_and_features,reverse=True)
print("Feature importance in descending order")
print("Values, Features")
for x, y in ranks_and_features:
    print(f"{x}, {y}")
```

```python
max_no_features =20
keys = [k[1] for k in ranks_and_features[:max_no_features] ] [::-1]
values = [k[0] for k in ranks_and_features[:max_no_features] ][::-1]

plt.barh(keys, values)
```

**Feature Impact**

```python
feature_impact = ranks_and_features
feature_impact.sort(key=lambda x: abs(x[0]), reverse=True)
#feature_importance_top_10 = feature_impact[:max_no_features]
print(pd.Series(feature_impact[:max_no_features]))
```
